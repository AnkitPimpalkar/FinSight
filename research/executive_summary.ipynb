````python
# Live Demo: Model Performance
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load actual results
raw_data = pd.read_csv('../artifacts/data_ingestion/raw_data.csv')
pred_df = pd.read_csv('../artifacts/model_prediction/predictions.csv')

# Extract key metrics
ticker = raw_data['Ticker'].iloc[0]
latest_price = raw_data['Close'].iloc[-1]
predicted_price = pred_df['predicted_close_price'].iloc[0]
prediction_error = abs(predicted_price - latest_price) / latest_price * 100

# Display key metrics
print("üéØ LIVE PREDICTION RESULTS")
print("=" * 40)
print(f"üìä Ticker: {ticker}")
print(f"üí∞ Latest Price: {latest_price:.2f}")
print(f"üîÆ Predicted Price: {predicted_price:.2f}")
print(f"üìà Prediction Error: {prediction_error:.1f}%")
print(f"üéØ Accuracy: {100 - prediction_error:.1f}%")
print("=" * 40)

# Quick visualization
plt.figure(figsize=(10, 6))
recent_prices = raw_data['Close'].tail(10)
days = range(len(recent_prices))

plt.plot(days, recent_prices, 'o-', label='Actual Prices', linewidth=2, markersize=8)
plt.axhline(y=predicted_price, color='red', linestyle='--', linewidth=2, 
            label=f'Next-Day Prediction: ${predicted_price:.2f}')
plt.scatter([len(days)], [predicted_price], color='red', s=100, zorder=5)

plt.title(f'{ticker} Price Prediction - {100-prediction_error:.1f}% Accuracy', 
          fontsize=14, fontweight='bold')
plt.xlabel('Days')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"‚úÖ Model successfully predicts {ticker} with {100-prediction_error:.1f}% accuracy!")
````
